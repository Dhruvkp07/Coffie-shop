##Finding the most selling Coffie

```python
most_selling_coffee = df['coffee_name'].value_counts().idxmax()
print(f"The most selling coffee is: {most_selling_coffee}")
```

##Finding the money earned by them

```python
coffee_sales = df.groupby('coffee_name')['money'].sum().sort_values(ascending=False)
display(coffee_sales)
```


Analyze the provided data to determine the most rushed and profitable time of day.
